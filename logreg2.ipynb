```python
# Importing libraries
# -------------------
import pandas as pd

# Importing the dataset
# ---------------------
dataset = pd.read_csv('http://bit.ly/kaggletrain')

cols = dataset.columns.tolist() # column names
cols = [cols[1]] + cols[0:1] + cols[2:]
dataset = dataset[cols]
dataset = dataset.drop(["Name", "Ticket", "Cabin"], axis = 1)

#see haed data
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
age_mean = dataset.Age.mean()

dataset.Age.fillna(value=age_mean, inplace = True)
```

```python
dataset.Embarked.fillna(value="S", inplace =True)
```

```python
#(shortcut) to code categorical varaibles
dataset = pd.get_dummies(dataset, columns = ["Sex", "Embarked"], drop_first = True)
```

```python
dataset.head()
```

```python
feature_cols = ['Pclass', 'Parch', 'Age']
```

```python
X = dataset.loc[:, feature_cols]
```

```python
X.shape
```

```python
y = dataset.Survived
```

```python
y.shape
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X,y)
```

```python
test = pd.read_csv('http://bit.ly/kaggletest')
```

```python
test.shape
```

```python
test.isnull().sum()
```

```python
test.Age.fillna(value=age_mean, inplace = True)
```

```python
X_new = test.loc[:, feature_cols]
```

```python
X_new.shape
```

```python
X_new.info()
```

```python
new_pred_class = logreg.predict(X_new)
```

```python
new_pred_class
```

```python
pd.DataFrame({'PassengerId' : test.PassengerId, 'Survived' : new_pred_class}).set_index('PassengerId').to_csv('titanic_pred_reg.csv')
```
